```python
# Extracting unique team IDs and their associated names from the original data

unique_teams = ranking[['TEAM_ID', 'TEAM']].drop_duplicates()


```

```python
# Grouping the data by 'TEAM_ID' and aggregating the other columns
grouped_data1 = teams.groupby('TEAM_ID').agg({
    'NICKNAME':'first'
})


```

```python
# Grouping the data by 'TEAM_ID' and aggregating the other columns
grouped_data = ranking.groupby('TEAM_ID').agg({
    'G': 'sum',  # Summing up games played
    'W': 'sum',  # Summing up wins
    'L': 'sum',  # Summing up losses
    'W_PCT': 'mean',  # Average win percentage
    'TEAM':'first'
})

# Sorting the grouped data based on the highest number of games played
grouped_data_sorted_by_games = grouped_data.sort_values(by='G', ascending=False)
result_df = pd.merge(grouped_data_sorted_by_games, grouped_data1, on='TEAM_ID', how='left')
# Display the resulting dataframe
result_df.head()

```

```python
# Concatenate 'TEAM' and 'NICKNAME' columns and store the result in a new column 'Team_Nickname'
result_df['Team_Nickname'] = result_df['TEAM'] + ' ' + result_df['NICKNAME']
result_df.head()
```

```python
# Dropping the 'TEAM' and 'NICKNAME' columns
result_df = result_df.drop(['TEAM', 'NICKNAME'], axis=1)

# Display the updated dataframe without the dropped columns
result_df.head()
```

```python
import matplotlib.pyplot as plt

# Assuming result_df is already defined and has the columns 'W', 'L', and 'Team_Nickname'

# Creating the subplots for the pie charts, adjusted for 6 teams
fig, axes = plt.subplots(1, 5, figsize=(20, 6))  # Adjusted layout for 6 pie charts
axes = axes.flatten()  # Flattening the axes array for easy iteration
plt.suptitle("Top 5 Teams with the Highest Number of Games Played", fontsize=16, x=0.4)

# Use similar shades for the pie charts
colors = ['#ABAAD0', '#8E8EAB']
legend_labels = ['Wins', 'Losses']

for i, (idx, row) in enumerate(result_df.iterrows()):
    # Check if the index is within the range of axes
    if i >= len(axes):
        break  # Break the loop if i exceeds the length of axes

    # Data for pie chart
    labels = 'Wins', 'Losses'
    sizes = [row['W'], row['L']]
    team_name = row['Team_Nickname']

    # Plotting the pie chart
    wedges, texts, autotexts = axes[i].pie(sizes, autopct='%1.1f%%', startangle=140, colors=colors)
    axes[i].set_title("")  # Clear the default title

    # Arrange title in center
    axes[i].pie([1], colors=['w'], radius=0.4)  # Create a white circle at the center
    for autotext in autotexts:
        autotext.set_fontsize(12)  # Set the fontsize for percentage labels
    axes[i].set_xlabel(team_name, fontsize=14, ha='center')

# Create a common legend outside the subplots
fig.legend(legend_labels, loc='upper right', bbox_to_anchor=(0.86, 0.86), fontsize=10)

# Ensure the loop doesn't exceed the axes array size
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to accommodate the main title and legend
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have already defined 'top_10_players' and 'top_10_players_with_highest_score' DataFrames

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Identify common player names between the two DataFrames
common_players = set(top_10_players['index']).intersection(top_10_players_with_highest_score['PLAYER_NAME'])

# Create a color list where common players have the same color and others are gray
colors1 = ['#7C7C8E' if player in common_players else '#D7D7E1' for player in top_10_players['index']]
colors2 = ['#7C7C8E' if player in common_players else '#D7D7E1' for player in top_10_players_with_highest_score['PLAYER_NAME']]

# First subplot for 'Games Played'
ax1.bar(top_10_players['index'], top_10_players['Count'], color=colors1)
ax1.set_ylabel('Games Played', fontsize=12)
ax1.set_xlabel('Player Name', fontsize=12)
ax1.set_title('Top 10 Players by Games Played', fontsize=14)
ax1.tick_params(axis='x', rotation=90)  # Rotate x-axis labels by 90 degrees

# Add data labels above the bars for the first subplot
for index, value in enumerate(top_10_players['Count']):
    ax1.text(index, value, str(value), ha='center', va='bottom', fontsize=10)

# Second subplot for 'Total Points Score'
ax2.bar(top_10_players_with_highest_score['PLAYER_NAME'], top_10_players_with_highest_score['PTS'], color=colors2)
ax2.set_ylabel('Total Points Score', fontsize=12)
ax2.set_xlabel('Player Name', fontsize=12)
ax2.set_title('Top 10 Players With Highest Point', fontsize=14)
ax2.tick_params(axis='x', rotation=90)  # Rotate x-axis labels by 90 degrees

# Add data labels above the bars for the second subplot
for index, value in enumerate(top_10_players_with_highest_score['PTS']):
    value_in_k = value / 1000  # Convert to K format
    ax2.text(index, value, f'{value_in_k:.2f}k', ha='center', va='bottom', fontsize=10)

# Customize the appearance of the grid lines (optional)

# Remove the top and right spines for both subplots (optional)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# Ensure proper spacing
plt.tight_layout()

# Display the subplots
plt.show()
```

# Graph 03


```python

grouped_data = games.groupby('SEASON').agg({
    'HOME_TEAM_WINS': 'sum',
    'SEASON': 'count'  # Count the number of rows (matches) for each season
})

# Rename the columns for clarity (optional)
grouped_data.rename(columns={'HOME_TEAM_WINS': 'TOTAL_HOME_TEAM_WINS', 'SEASON': 'TOTAL_MATCHES'}, inplace=True)

# Display the result


grouped_data['TOTAL_AWAY_TEAM_WINS'] =grouped_data['TOTAL_MATCHES']-grouped_data['TOTAL_HOME_TEAM_WINS']

grouped_data['PERCENTAGE_HOME_TEAM_WINS'] = round((grouped_data['TOTAL_HOME_TEAM_WINS'] / grouped_data['TOTAL_MATCHES']) * 100, 2)


grouped_data['PERCENTAGE_AWAY_TEAM_WINS'] = round((grouped_data['TOTAL_AWAY_TEAM_WINS'] / grouped_data['TOTAL_MATCHES']) * 100, 2)



grouped_data.reset_index(inplace=True)
grouped_data.head()


# Extracting the percentage of wins and the seasons
home_team_wins = grouped_data['PERCENTAGE_HOME_TEAM_WINS'].values
away_team_wins = grouped_data['PERCENTAGE_AWAY_TEAM_WINS'].values
seasons = grouped_data['SEASON'].astype(str).values  # Convert season numbers to strings for labeling purposes

# Generate y positions for each season
y_pos = np.arange(len(seasons))

# Increase the bar width
bar_width = 0.6

# Define shades of blue for the home and away team bars
color_home = '#C9C9DC'  # A lighter shade for the home team
color_away = '#ABAAD0'  # A bit darker shade for the away team

# Create the figure and axis objects again with the new bar width and colors
fig, ax = plt.subplots(figsize=(14, 8))

# Plot the bars for home team wins and away team wins in opposite directions with new colors
bars_home = ax.barh(y_pos, home_team_wins, height=bar_width, color=color_home, label='Home Team Wins')
bars_away = ax.barh(y_pos, -away_team_wins, height=bar_width, color=color_away, label='Away Team Wins')

# Add the season labels on the y-axis
ax.set_yticks(y_pos)
ax.set_yticklabels(seasons)
ax.invert_yaxis()  # Labels read top-to-bottom

# Adding data labels on the bars
for bar in bars_home:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height() / 2, f'{width}%', ha='right', va='center')

for bar in bars_away:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height() / 2, f'{-width}%', ha='left', va='center')

# Set labels, title, and legend
ax.set_xlabel('Percentage')
ax.set_title('Home and Away Team Wins Percentage by Season')
ax.legend(loc='upper right', bbox_to_anchor=(1.15, 1))

# Removing the spines and setting grid
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Display horizontal grid lines
ax.xaxis.grid(True)
ax.yaxis.grid(False)

# Set the x-axis limits to the specified range
ax.set_xlim(-50, 70)

# Remove x-axis ticks
ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# Remove grid lines
ax.grid(False)

# Show the plot with the new adjustments
plt.tight_layout()
plt.show()
```
